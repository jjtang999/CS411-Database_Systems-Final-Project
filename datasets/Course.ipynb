```python
import pandas as pd
```

```python
split_courses = recent_courses.copy()
```

```python
# Reformat to schema
split_courses['Subject Code'] = split_courses['Course'].str.split(expand=True)[0]
split_courses['Number'] = split_courses['Course'].str.split(expand=True)[1]
split_courses.rename({'Course Title': 'Name'})
```

```python
split_courses
```

```python
cols = ['Subject Code', 'Number', 'Abbreviation']
```

```python
def convert(col_name):
    return split_courses[split_courses[col_name].notnull()].reset_index().rename(columns={col_name: 'Abbreviation'})[cols]

acp = convert('ACP')
cs = convert('CS')
hum = convert('HUM')
nat = convert('NAT')
qr = convert('QR')
sbs = convert('SBS')
```

```python
def compute_year_term(year, term, year_term):
    raw_gpa_courses = pd.read_csv('raw/uiuc-gpa-dataset.csv')
    raw_gpa_courses = raw_gpa_courses.drop(columns=['YearTerm', 'Sched Type'])
    curr_raw_gpa_courses = raw_gpa_courses[(raw_gpa_courses['Year'] == year) & (raw_gpa_courses['Term'] == term) & raw_gpa_courses['Primary Instructor'].notnull()].reset_index().drop(columns=['index'])
    
    raw_course_offerings = pd.read_csv(f'raw/course-explorer/{year_term}.csv')
    
    filtered_offerings = raw_course_offerings[raw_course_offerings['Instructors'].notnull()]
    attrs = ['Year', 'Term', 'Subject', 'Number', 'Name', 'Description', 'Credit Hours', 'Start Time', 'End Time', 'Days of Week', 'Room', 'Building', 'Instructors']
    filtered_offerings[attrs]
    
    curr_raw_gpa_courses['CRN'] = ['hi'] * len(curr_raw_gpa_courses)
    curr_raw_gpa_courses['Description'] = ['hi'] * len(curr_raw_gpa_courses)
    curr_raw_gpa_courses['Section'] = ['hi'] * len(curr_raw_gpa_courses)
    curr_raw_gpa_courses['Start Time'] = ['hi'] * len(curr_raw_gpa_courses)
    curr_raw_gpa_courses['End Time'] = ['hi'] * len(curr_raw_gpa_courses)
    curr_raw_gpa_courses['Days of Week'] = ['hi'] * len(curr_raw_gpa_courses)
    curr_raw_gpa_courses['Room'] = ['hi'] * len(curr_raw_gpa_courses)
    curr_raw_gpa_courses['Building'] = ['hi'] * len(curr_raw_gpa_courses)
    curr_raw_gpa_courses['Credit Hours'] = ['hi'] * len(curr_raw_gpa_courses)
    
    other_index = last_good_index = 0
    invalids = []
    for i, row in curr_raw_gpa_courses.iterrows():
        other_index = last_good_index
        last_name = row['Primary Instructor'].split(',')[0]
        while other_index < len(filtered_offerings):
            if row['Year'] == filtered_offerings.iloc[other_index]['Year'] and row['Term'] == filtered_offerings.iloc[other_index]['Term'] and any([s.startswith(last_name) for s in filtered_offerings.iloc[other_index]['Instructors'].split(';')]):
                break
            other_index += 1
        else:
#             print(f'Error on {i}, {year}:{term}')
            invalids.append(i)
            continue
        curr_raw_gpa_courses.loc[i, 'CRN'] = filtered_offerings.iloc[other_index]['CRN']
        curr_raw_gpa_courses.loc[i, 'Description'] = filtered_offerings.iloc[other_index]['Description']
        curr_raw_gpa_courses.loc[i, 'Section'] = filtered_offerings.iloc[other_index]['Section']
        curr_raw_gpa_courses.loc[i, 'Start Time'] = filtered_offerings.iloc[other_index]['Start Time']
        curr_raw_gpa_courses.loc[i, 'End Time'] = filtered_offerings.iloc[other_index]['End Time']
        curr_raw_gpa_courses.loc[i, 'Days of Week'] = filtered_offerings.iloc[other_index]['Days of Week']
        curr_raw_gpa_courses.loc[i, 'Room'] = filtered_offerings.iloc[other_index]['Room']
        curr_raw_gpa_courses.loc[i, 'Building'] = filtered_offerings.iloc[other_index]['Building']
        curr_raw_gpa_courses.loc[i, 'Credit Hours'] = filtered_offerings.iloc[other_index]['Credit Hours']
        last_good_index = other_index

    return curr_raw_gpa_courses.drop(index=invalids)
```

```python
entries = [(2021, 'Spring', '2021-sp'), (2020, 'Fall', '2020-fa'), (2020, 'Spring', '2020-sp'), (2019, 'Fall', '2019-fa'), (2019, 'Spring', '2019-sp'), (2018, 'Fall', '2018-fa'), (2018, 'Spring', '2018-sp'), (2017, 'Fall', '2017-fa'), (2017, 'Spring', '2017-sp')]
end = pd.DataFrame([], columns=['Year', 'Term', 'Subject', 'Number', 'Course Title', 'A+', 'A', 'A-',
       'B+', 'B', 'B-', 'C+', 'C', 'C-', 'D+', 'D', 'D-', 'F', 'W',
       'Primary Instructor', 'CRN', 'Description', 'Section', 'Start Time',
       'End Time', 'Days of Week', 'Room', 'Building', 'Credit Hours'])
for y,t,yt in entries:
    curr = compute_year_term(y,t,yt)
    end = pd.concat([end, curr])
```

```python
offerings_cols = ['Year', 'Term', 'Subject', 'Number', 'A+', 'A', 'A-',
       'B+', 'B', 'B-', 'C+', 'C', 'C-', 'D+', 'D', 'D-', 'F', 'W',
       'Primary Instructor', 'CRN', 'Section', 'Start Time',
       'End Time', 'Days of Week', 'Room', 'Building']
offerings = end[offerings_cols]
offerings.to_csv('processed/offerings.csv', index=False)
```
