# Note

Bruh this data is so badly formatted. Why so many duplicate entries for the same person and the same title????????
This makes matching the prof to a department a bit trickier, prolly just going to match to the first department we can find.

- Going to remove duplicate entires in the uniques table since those are probably different people
- Going to take the first entry in the duplicates table

```python
real_uniques = uniques[~uniques['Employee Name'].isin(['Liu, Wei', 'Thomas, Susan Mary'])]
print(len(real_uniques))
print(len(set(real_uniques['Employee Name'])))
```

```python
duplicates.groupby(['Employee Name'], group_keys=True).first().reset_index()
```

```python
real_duplicates = duplicates.groupby(['Employee Name'], group_keys=True).first().reset_index()
print(len(real_duplicates))
print(len(set(real_duplicates['Employee Name'])))
```

```python
processed_faculty = pd.concat([real_uniques, real_duplicates])
```

```python
print(len(processed_faculty))
print(len(set(processed_faculty['Employee Name'])))
```

```python
colleges = processed_faculty.groupby(['College Code']).first().reset_index()[['College Code', 'College Name']]
departments = processed_faculty.groupby(['Department Code']).first().reset_index()[['Department Code', 'Department Name']]
```
